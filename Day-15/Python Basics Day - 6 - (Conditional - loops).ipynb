# Itaretive Statements

## While loop
- repeatedly executes a target statement as long as a given condition
is true.

```python
while True:
    print(x)
    x -= 1
```

## for loop
- It has the ability to iterate over the items of any sequence, such as a list or
a string


```python
bepec = [1,2,3,4,5]
for i in bepec:
    print(i)
```

```python
for x, y in enumerate(bepec):
    print(x,y)
```

```python
dict1 = {1:2, 3:4}

for i,j in dict1.items():
    print(i,j)
```

# Transfer Statements

## Break

```python
# Break Statement: It brings control out of the loop
for i in range(1,10):
    if i == 6:
        break
    print(i)
```

## continue

```python
# Continue Statement: It returns the control to the beginning of the loop
for i in range(1,10):
    if i == 6:
        continue
    print(i)
```
