# Escape sequences are control character used to move the cursor and print characters such as ',".\ and so on.

```python
print("Hello 'world'")
```

```python
print("Hello \"world\"")
```

```python
b = "This is \\ backslash."
print(b) 
```

```python
b = "Welcome\neveryone\nhappy\nlearning"
print(b)
```

```python
b = "Welcome\teveryone"
print(b) 
```

```python
b = "Welcome  \beveryone"
print(b)
```

```python
print('hi welcome to \r bepec')
```
