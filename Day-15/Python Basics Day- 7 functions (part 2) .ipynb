# Function Arguments in Python

# positional (args)

```python
def student(name, age):
    print("Name :", name)
    print("Age :", age)
    
student('john', 21)
```

```python
student(21, "john" )
```

# keyword (args)

```python
def student(name, age):
    print("Name :", name)
    print("Age :", age)
```

```python
student(age= 21, name = 'john', )
```

# Default (arg)

```python
def student(name, age, company='xyz'):
    print("Name :", name)
    print("Age :", age)
    print("Company :", company)
```

```python
student(name='john',age= 28)
```

```python
student(name= 'harsha', age= 25, company='yzx')
```

#  (arg,*args)

- The special syntax *args is used to pass a non-keyworded, variable-length argument.

```python
def student(name, age,*company):
    print('name:', name)
    print("Age:", age)
    print("company : ", company)
    
```

```python
student("rahul", 25, "bepec", "xyz", "yzx")
```

```python
student(name = 'anurag', age = 26, company1= 'xyz', company2= 'yzx', company3 = 'abc')
student(name = 'r', age = 26, company1= 'xyz', company2= 'yzx', company3 = 'abc')
```

```python
def test(*arg):
    for i in arg:
        print(i)
        
test(1,2,3,4,5)
```

```python
def add(name,*marks):
    total = 0
    for i in marks:
        total = total+i
    print(f'{name}: {total}')
    
add('rohith',10,20)
add('shashi', 10,30,40,50)
add('akhil',1000,50,30,100,10,10,10,10.1)
add('koti', 100,500,1000)
```

```python
def student(name, age, *company):
    print("Name :", name)
    print("Age :", age)
    print("Company :", company)
```

```python
student('john', 21, 'bepec', 'xyz', 'yzx')
```

# (arg, **kwargs)
- The special syntax **kwargs is used to pass a keyworded, variable-length argument. The reason is because the double star allows us to pass through keyword arguments (and any number of them)

```python
def student(name, age, **company):
    print("Name :", name)
    print("Age :", age)
    print("Company :", company)
```

```python
student(name='john', age=21, company1='bepec',company2='xyz',company3='yzx')
```

```python
def student(name, age, **company):
    print("Name :", name)
    print("Age :", age)
    
    for i,j in company.items():
        print(i,":",j)
```

```python
student(name='john', age=21, company1='bepec',company2='xyz',company3='yzx')
```
