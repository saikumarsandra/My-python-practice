```python
import numpy as np
x =np.arange(0,100,2)
x1=np.arange(10,50,1)
print(x1)

print(x)
```

```python
y=x*2
y1=x1**3
print(y)
print(y1)
```

```python
plt.plot(x,y,'g-',label='x vs y')
plt.plot(x1,y1,'r--',label='x1 vs y1')
plt.xlabel('x,x1')
plt.ylabel('y,y1')
plt.legend()
plt.show()
```

```python
name=['group_a','group_b','group_c']
values=[1,10,10]
plt.figure(figsize=(10,5))

plt.subplot(131)
plt.bar(name,values)
plt.subplot(132)
plt.scatter(name,values)
plt.subplot(133)
plt.plot(name,values)
plt.suptitle('categorical values')
plt.show()
```
