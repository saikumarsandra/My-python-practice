```python
#import the required libraries
import pandas as pd 
import seaborn as sns 
import numpy as np 
```

```python
#import tips dataset from seaborn 
tip = sns.load_dataset("tips")
```

```python
#get the shape of the data
tip.shape
```

```python
#view the top 20 row 
tip.head(20)
```

```python
#view the bottom 20 row 
tip.tail(20)
```

```python
#Get the avg total_bill by only male
tip.loc[tip['sex'] == 'Male', 'total_bill'].mean()
```

```python
#Get the avg(mean,median) total_bill and tip given by smoker and non smoker
pd.pivot_table(data= tip, values=['total_bill', 'tip'], index= ['smoker'], aggfunc= ['mean', 'median'])
```

```python
# what is most frequent time of vist
tip['time'].value_counts()
```

```python
#Which day customers are giving higest tips
tip['tip'].groupby(tip.day).mean()
```

```python
#get the avg(median) tip and total_bill amount by smoker(yes) on sunday
pd.pivot_table(data= tip, values=['total_bill', 'tip'], index= ['smoker', 'day'], aggfunc= [ 'median'])
```

```python
#get the (mean,median,min,max) tip and total_bill amount by smoker(yes) on friday with time(Lunch)
#1 What is min bill by a smoker on sunday
#2 what is max tip by a non smoker on thursday
#3 what is total_bill made on sat by non smokers

pd.pivot_table(data= tip, values=['total_bill', 'tip'], index= ['smoker', 'day', 'time'], aggfunc= [ 'mean', 'median', 'min', 'max'])
```

```python
#which gender is having highest bill amount
tip[['sex', 'total_bill']].groupby('sex').mean()
```

```python
#convert columns to rows 
pd.melt(tip)
```

```python
#Get the min, max, and avg(mean, and median) of total bill and tip
tip[['total_bill', 'tip']].describe()
```

```python
tip.describe()
```

```python
#create a new column total_bill amount category(low, moderate, high)
tip['total_bill_class'] = tip['total_bill'].apply(lambda x: 'low' if x< 15
                                       else('moderate' if 15<=x<35
                                       else'high'))
```
