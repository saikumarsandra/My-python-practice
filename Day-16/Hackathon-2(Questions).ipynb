```python
#Get the avg total_bill by only male
AvgTotal=tip.groupby(by='sex')['total_bill'].mean()
print("avarage total bill by male :",AvgTotal.Male)
```

```python
#Get the avg(mean,median) total_bill and tip given by smoker and non smoker
mean1=tip[tip['smoker']=='Yes']['total_bill'].mean()
mean2=tip[tip['smoker']=='No']['total_bill'].mean()
median1=tip[tip['smoker']=='Yes']['tip'].median()
median2=tip[tip['smoker']=='No']['tip'].median()
print(mean1,mean2,median1,median2)
avarage = (mean1+mean2+median1+median2)/4
print("avarage ( mean ,median):",avarage)
```

```python
# what is most frequent time of vist
frequentTime= tip['time'].value_counts()
print(frequentTime)
if frequentTime.Dinner > frequentTime.Lunch:
    print("most frequent visited time is Dinner")
else:
    print("most frequent visited time is Lunch")
```

```python
#Which day customers are giving higest tips
day = tip['day'].unique()
tip1= []

for d in day:
    tip1.append(tip[tip['day']==d]['tip'].max())
    
for i,j in zip(day,tip1):
    print(i,"=",j)
```
