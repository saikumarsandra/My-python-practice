```python
import pandas as pd 
import numpy as np 
import matplotlib
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv(r'C:\Users\saileo\Downloads\ALL_ROOM_BILLS.csv')
```

```python
x = dataset['TOTAL']
y = dataset['Groceries']
plt.figure(figsize=(8,10))
plt.title('Grocery and Total')
plt.xlabel('Total amount')
plt.ylabel('Grocery names')
plt.plot(x,y,'g+--')
plt.show()
```

```python
dataset.boxplot(column=['SAI KUMAR','PAVAN','GOPI'])
```

```python
dataset.hist(bins=20)
```

```python
names = dataset['Groceries']=='Global'
values = dataset['TOTAL']

plt.figure(figsize=(15, 8))

plt.subplot(131)
plt.bar(names, values)
plt.xlabel('Global bill')
plt.ylabel('Global grocery')
plt.subplot(132)
plt.scatter(names, values)
plt.xlabel('Global bill')
plt.ylabel('Global grocery')
plt.subplot(133)
plt.grid()
plt.plot(names, values,'*-')
plt.xlabel('Global bill')
plt.ylabel('Global grocery')
plt.suptitle('Categorical Plotting')
plt.show()
```
