# Basic Visualization Rules

1. __Type of plot__ : The first step is to choose the appropriate plot type. 
2. __Axes formation__ : Second, when we choose type of plot, one of the most important things is to label your axis. If we don’t do this, the plot is not informative enough. 
3. __Title__ : we can add a title to make our plot more informative.
4. __Labels__: add labels for different categories when needed.
5. __Annotation, optionally we can add a text or an arrow at interesting data points.
6. __ Figure sie, color etc__ in some cases we can use some sizes and colors of the data to make the plot more informative.

# Basic matplotlib

> - import matplotlib
> - import matplotlib.pyplot as plt
> - from matplotlib import style
> - %matplotlib inline 

- __plt.title__(“My Title”) will add a title “My Title” to your plot
- __plt.xlabel__(“Year”) will add a label “Year” to your x-axis
- __plt.ylabel__(“Population”) will add a label “Population” to your y-axis
- __plt.xticks__([1, 2, 3, 4, 5]) set the numbers on the x-axis to be 1, 2, 3, 4, 5. We can also pass and labels as a second argument. For, example, if we use this code plt.xticks([1, 2, 3, 4, 5], ["1M", "2M", "3M", "4M", "5M"]), it will set the labels 1M, 2M, 3M, 4M, 5M on the x-axis.
- __plt.yticks__() - works the same as plt.xticks(), but for the y-axis.

# import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# subplots

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 10, 12, 3])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

# lineplot

```python
plt.plot([1,2,3,4,5], [2,3,4,5,6])
plt.title("test plot")
plt.xlabel("Xlable")
plt.ylabel("ylable")
plt.show()
```

```python
plt.plot([1,2,3,4,5], "g",[2,3,4,5,6,7], 'bo' ,markersize= 20, linewidth =20 )
```

```python
a = np.arange(0,10)
plt.plot(a,'r+:', a**2, 'g^--', a**3, "b>-" )
```

```python
plt.plot([10,8,6,4,2], "r+:", markersize=20, linewidth=4, markeredgecolor="green", markeredgewidth = 10, alpha=0.2)
```

# X&y label

```python
plt.plot([6,5,4,3,2,1])
plt.xlabel("Xlabel", fontdict={'size':20, 'color':'red', 'weight':'black'}, labelpad=4, loc='left')
plt.show()
```

# title

```python
plt.plot([6,5,4,3,2,1], color='orange' ,marker ='+', markersize= 20)
plt.title("test", fontdict={'size':40, 'color':'gray'}, pad=4, loc='left', weight='black')
```

# Grid

```python
plt.plot([6,5,4,3,2,1], color='orange')
plt.title("test", fontdict={'size':40, 'color':'gray'}, pad=4, loc='left', weight='black')
plt.minorticks_on()
plt.grid(b=True,which='minor' ,color= 'red')
plt.grid(b=True, which= 'major', color ='blue')
```

```python
plt.plot([6,5,4,3,2,1], color='orange')
plt.title("test", fontdict={'size':40, 'color':'gray'}, pad=4, loc='left', weight='black')
plt.minorticks_on()
plt.grid(b=True, which= 'minor', color ='blue' ,axis= 'y')
```
