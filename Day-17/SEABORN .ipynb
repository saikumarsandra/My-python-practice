```python
import pandas as pd 
import numpy  as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
data= sns.load_dataset('tips')
```

```python
data.head(10)
```

<h2>relplot</h2>

```python
sns.relplot(data = data , x ='total_bill',y= 'tip' ,col='sex',hue='smoker')
```

```python
sns.relplot(data = data , x ='total_bill',y= 'tip',hue='smoker', style='sex')
```

```python
sns.relplot(data = data , x ='total_bill',y= 'tip',hue='smoker',col='sex', row='smoker',style='sex')
```

```python
sns.relplot(data = data , x ='total_bill',y= 'tip',hue='smoker',col='sex', row='smoker',style='sex',palette='Blues')
```

distplot

```python
sns.distplot(data['tip'])
```

```python
sns.distplot(data['tip'])
```

```python
sns.distplot(data['tip'],kde=False,hist=True)
```

```python
sns.distplot(data['tip'],kde=True,hist=False)
```

```python
sns.distplot(data['tip'],kde=True,hist=False,vertical=True)
```

```python
sns.distplot(data['tip'],kde=True,hist=True,bins=80)
```

Barplot

```python
sns.barplot(data=data, x='day',y='tip', estimator=np.min)
```

countplot

```python
sns.countplot(data=data, x='sex',hue='time')
```

```python
sns.countplot(data=data, y='sex',hue='time')
```

Boxenplot

```python
sns.boxenplot(data=data,x='day',y='tip')
```

violinplot

```python
sns.violinplot(data=data,x='day',y='tip',hue='sex')
```

# jointplot

```python
sns.jointplot(data=data,x='day',y='tip')
```

# pairplot

```python
sns.pairplot(data=data)
```

# heatmap

```python
sns.heatmap(data=data.corr(),annot=True,cmap='Set1')
```
