```python
plt.figure(figsize=(10,10))
sns.catplot(y="age" ,hue="gender", kind="count",data=dataset,edgecolor='0.5',legend=False)
plt.legend({'1= male','0= female'})
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(dataset.corr(),annot=True,fmt='.2g',cmap='Set1')
plt.show()
```

```python
sns.pairplot(dataset)
```

```python
sns.scatterplot(data = dataset , x = 'chol',y='age',hue='cp')
plt.xlabel('Chest pain Type :1.typical angina,2.atypical angina,3.non-anginal pain,4.asymptomatic')
plt.ylabel('age')

plt.show()

```

```python
sns.lineplot(data = dataset , x = 'chol',y='trestbps',hue='gender')
plt.xlabel('cholestrol')
plt.ylabel('Resting blood pressure')
plt.legend({'1= male','0= female'})
plt.show()
```
