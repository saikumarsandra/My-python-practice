```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

# relplot

```python
sns.relplot(data= data, x='total_bill', y= 'tip', col='sex', hue='smoker', row='smoker', palette='flare')
```

```python
sns.scatterplot(data= data, x='total_bill', y= 'tip', hue='time', palette='Blues')
```

# Distplot

```python
sns.distplot(data['tip'], kde=True, hist= True, color='r', vertical=False,bins=10)
```

# barplot

```python
sns.barplot(data= data, x='day', y='tip', estimator=np.max)
```

# countplot

```python
sns.countplot(data= data, x='sex', hue='time')
```

# boxenplot

```python
sns.boxenplot(data= data, x='day', y='tip')
```

# violinplot

```python
sns.violinplot(data= data, x='day', y='tip', hue='sex')
```

# jointplot

```python
sns.jointplot(data=data, x='total_bill', y='tip', color='gray', kind= 'kde')
```

# pairplot

```python
sns.set_style("dark")
sns.pairplot(data)
```

# heatmap

```python
sns.heatmap(data=data.corr(), annot=True, cmap='Set1')
```

# facetgrid

```python
b = sns.FacetGrid(data, col='day')
b.map(plt.hist, 'total_bill')
```
