```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as ply
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
dataset = sns.load_dataset('tips')
```

```python
heart=dataset.head(20)
heart
```

```python
import plotly.io as pio
pio.templates
```

```python
ply.bar(heart,x= 'sex', y='total_bill', labels={'sex':'Gender'},color="tip",hover_data=['smoker', 'size'])
```

```python
fig=ply.histogram(heart,x='total_bill',color='sex',labels={'sex':'Gender'},marginal="violin")
fig.show()
```

```python
data = dict(
    number=heart['total_bill'],
    stage=heart['sex'])
fig = ply.funnel(data, x='number', y='stage',color='stage')
fig.show()
```

```python
fig = ply.pie(heart, values='tip',names='sex',color_discrete_sequence=ply.colors.sequential.RdBu)
fig.show()
```

```python
fig = ply.sunburst(dataset, path=['total_bill', 'day','time'], values='tip',color='time',color_discrete_map={'(?)':'black', 'Lunch':'gold', 'Dinner':'darkblue'})
fig.show()
```

```python
fig = ply.treemap(dataset, path=['day', 'time', 'sex'], values='total_bill')
fig.show()
```

```python
fig = ply.scatter_3d(heart, x='tip', y='total_bill', z='size',
                     symbol='day')
fig.show()
```

```python
fig = ply.box(dataset, x="day", y="total_bill", color="smoker")
fig.update_traces(quartilemethod="exclusive") 
# or "inclusive", or "linear" by default
```

```python

fig = ply.violin(dataset, y="tip", x="smoker", color="sex", box=True, points="all",
          hover_data=['day'])
fig.show()
```

```python
fig = ply.density_heatmap(dataset, x="total_bill", y="tip", marginal_x="histogram", marginal_y="violin")
fig.show()
```
